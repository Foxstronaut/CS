### 5. Complete the code:

````java
import java.util.Scanner;

public class PromptNumbers {
    public static void main(String[] args) {
        Scanner console = new Scanner(System.in);
        System.out.print("How many numbers will you enter? ");
        int count = console.nextInt();

        int[] ctnr = new int[count];

        for (int i = 0; i < count; i++) {
            System.out.print("Type a number: ");
            ctnr[i] = console.nextInt();
        }

        System.out.println();
        System.out.println("Your numbers in forward order:");
        for (int i = 0; i < count; i++) {
            System.out.println(ctnr[i]);
        }

        System.out.println();
        System.out.println("Your numbers in backward order:");
        for (int i = 0; i < count; i++) {
            System.out.println(ctnr[count - 1 - i]);
        }
    }
}
````

### 6.

````java
// Reads temperatures from the user, computes average and # days above average
import java.util.*;

public class Weather {
   public static void main(String[] args) {
      Scanner console = new Scanner(System.in);
      System.out.print("How many days' temperatures? ");
      int days = console.nextInt();

      int[] temps = new int[days];   // store each day's temp
      int sum = 0;

      for (int i = 0; i < days; i++) {
         System.out.print("Day " + (i + 1) + "'s high temp: ");
         temps[i] = console.nextInt();
         sum += temps[i];
      }

      // Compute and print out the overall average
      double average = (double) sum / days;
      System.out.printf("Average temp = %.1f%n", average);

      // Compute and print out the number of days above average
      int aboveAvg = 0;
      for (int i = 0; i < days; i++) {
         if (temps[i] > average) {
            aboveAvg++;
         }
      }
      System.out.println(aboveAvg + " day(s) above average");
   }
}
````

### 7. Examine a large integer and return an array that stores the count of the number of occurrences of every digit from 0 through 9.
 
For example: 
 
The number 229231007 contains two 0s, one 1, three 2s, one 7, and one 9

````java
public class DigitCounter {

    public static int[] digitCounter(int number) {
        int[] counts = new int[10];

        if (number == 0) {
            counts[0] = 1;
        } else {
            int temp = number;
            while (temp > 0) {
                counts[temp % 10]++;
                temp /= 10;
            }
        }

        return counts;
    }

    
}
````

now here's a demo:

````java
Scanner scanner = new Scanner(System.in);
System.out.print("Enter a large integer: ");
int number = scanner.nextInt();
scanner.close();

int[] result = DigitCounter.digitCounter(number);

for (int i = 0; i < result.length; i++) {
    if (result[i] > 0) {
        System.out.println(i + " appears " + result[i] + " time(s)");
    }
}
````
